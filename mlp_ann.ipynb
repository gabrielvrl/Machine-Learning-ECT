```
# Artificial Neural Network
# from https://www.superdatascience.com/machine-learning/ 


# Part 1 - Data Preprocessing

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Mais sobre pandas: https://github.com/Natalnet/GCiD/blob/master/Codes/Basic%20Module/01%20-%20Introdu%C3%A7%C3%A3o%20%C3%A0%20Leitura%20e%20Visualiza%C3%A7%C3%A3o%20de%20Dados.ipynb 
```

```
# Importing the dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/ect-info/ml/master/dados/Churn_Modelling.csv')


dataset.head( 10 )
```

```
dataset.describe()
```

```
X = dataset.iloc[:,[3,6,7,8,9,10,11,12]].values
y = dataset.iloc[:, 13].values

print(X[0:6,:])
```

```
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.05, random_state = 0)
```

```
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
print(X_test[0:6,:])
```

```
# Part 2 - Now let's make the ANN!

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
classifier = Sequential()
```

```
# Adding the input layer and the first hidden layer
classifier.add(Dense( activation = 'relu', input_dim = 8, units = 4, kernel_initializer = 'uniform'))


# Adding the second hidden layer
classifier.add(Dense( activation = 'relu', units = 6, kernel_initializer = 'uniform' ))


# Adding the second hidden layer
classifier.add(Dense( activation = 'relu', units = 10, kernel_initializer = 'uniform' ))



# Adding the output layer
classifier.add(Dense( activation = 'sigmoid', units = 1, kernel_initializer = 'uniform'))
```

```

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 50)
```

```
# Part 3 - Making the predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print(y_pred[0:10])

y_pred = (y_pred > 0.5)
print(y_pred[0:10])

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```
print("Matriz de Confusão:")
print(cm)
print("Taxa de acerto:")
print((cm[0,0]+cm[1,1])/len(y_test) )
print(len(y_test))


```
